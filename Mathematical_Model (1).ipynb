```python
def is_top(values):
  top_points = []
  for index, row in values.iterrows():
    target = float(row[5])
    avg_of_first = sum(float(x) for x in row[:5]) / 5
    avg_of_second = sum(float(x) for x in row[6:]) / 4
    if np.logical_and(avg_of_first < target, avg_of_second < target):
      top_points.append(1)
    else:
      top_points.append(0)
  count=0
  for v in top_points:
    if(v == 1):
      count+=1
  print('all points:', len(top_points))
  print('top points: ', count)
  return top_points
```

```python
values = df.iloc[:,:]
```

```python
values
```

```python
top_points = is_top(values)
```

Writing to CSV

```python
data_list_with_header = [['Turning points']] + [[value] for value in top_points]

# Writing to the CSV file
with open('aapl_M_model_output', 'w', newline='') as csvfile:
    # Create a CSV writer object
    csv_writer = csv.writer(csvfile)

    # Write the data list with column name to the CSV file
    csv_writer.writerows(data_list_with_header)

print(f'Data has been written')
```
